```python
# ------------------------------------------------------------
# CNN vs DNN for RIS Pixel Failure Detection
# Full pipeline: Data generation -> DNN baseline -> CNN (U-Net) ->
#   Training -> Evaluation -> SINR comparison (simple model)
# Paste into Google Colab / local Jupyter and run cell-by-cell.
# ------------------------------------------------------------

# 0. Optional: use this to mount Google Drive (if you want to save models)
# from google.colab import drive
# drive.mount('/content/drive')

# 1. INSTALL / IMPORT
!pip install -q tensorflow scikit-learn matplotlib seaborn

import os
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import layers, Model, Input
from tensorflow.keras.layers import Conv2D, MaxPooling2D, UpSampling2D, Concatenate, BatchNormalization, Activation, Dropout, Dense, Flatten
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

```

```python
# Reproducibility seeds
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)
random.seed(SEED)

# 2. SETTINGS
SAMPLES = 40000        # total samples (reduce if memory constrained)
RIS_SIZE = 12          # 12x12 RIS
CHANNELS = 2           # real and imag channels
TEST_SIZE = 0.2        # final test split
VAL_SPLIT = 0.1        # validation during training
BATCH_SIZE = 64
EPOCHS = 80

# Failure model parameters
MEAN_FAIL_RATE = 0.12   # average fraction of failed pixels per sample
STD_FAIL_RATE = 0.05

# 3. DATASET GENERATION: realistic per-pixel multiplicative failures
def generate_dataset(n_samples=SAMPLES, ris_size=RIS_SIZE,
                     mean_fail=MEAN_FAIL_RATE, std_fail=STD_FAIL_RATE,
                     complex_channels=CHANNELS, seed=SEED):
    np.random.seed(seed)
    X = np.zeros((n_samples, ris_size, ris_size, complex_channels), dtype=np.float32)
    Y = np.zeros((n_samples, ris_size, ris_size, 1), dtype=np.uint8)  # per-pixel failure label

    for i in range(n_samples):
        # baseline complex reflection (real, imag)
        real = np.random.normal(0, 1, (ris_size, ris_size))
        imag = np.random.normal(0, 1, (ris_size, ris_size))

        # sample failure probability for this sample
        p_fail = float(np.clip(np.random.normal(mean_fail, std_fail), 0.01, 0.8))
        fail_mask = (np.random.rand(ris_size, ris_size) < p_fail).astype(np.uint8)
        Y[i, :, :, 0] = fail_mask

        # draw failure parameters for failed pixels
        kappas = np.random.uniform(0.0, 1.0, (ris_size, ris_size))
        phases = np.random.uniform(-np.pi, np.pi, (ris_size, ris_size))
        z_real = kappas * np.cos(phases)
        z_imag = kappas * np.sin(phases)

        # apply multiplicative effect only where failed
        # complex multiplication: (a+jb)*(c+jd) = (ac - bd) + j(ad + bc)
        R = real.copy()
        I = imag.copy()
        idx = np.where(fail_mask == 1)
        if idx[0].size > 0:
            r_coords, c_coords = idx
            for (r, c) in zip(r_coords, c_coords):
                a = real[r, c]; b = imag[r, c]
                c_ = z_real[r, c]; d_ = z_imag[r, c]
                R[r, c] = a * c_ - b * d_
                I[r, c] = a * d_ + b * c_
        # assign channels
        X[i, :, :, 0] = R
        X[i, :, :, 1] = I

    return X, Y

print("Generating dataset... (this may take ~20-40s depending on SAMPLES)")
X, Y = generate_dataset()
print("X shape:", X.shape, "Y shape:", Y.shape)
# quick check of average failure rate
```

```python
avg_fail = Y.mean()
print("Average per-pixel failure rate (dataset-wide): {:.3f}".format(avg_fail))

# 4. Train / Validation / Test split
X_train_full, X_test, Y_train_full, Y_test = train_test_split(X, Y, test_size=TEST_SIZE, random_state=SEED)
# we will use validation via validation_split in model.fit or split further
print("Train:", X_train_full.shape, Y_train_full.shape, "Test:", X_test.shape, Y_test.shape)

```

```python
# 5. Normalization: compute per-channel mean/std on training set
channel_means = X_train_full.mean(axis=(0,1,2))
channel_stds = X_train_full.std(axis=(0,1,2)) + 1e-9
print("channel_means:", channel_means, "channel_stds:", channel_stds)

X_train_full = (X_train_full - channel_means) / channel_stds
X_test = (X_test - channel_means) / channel_stds

```

```python
# 6. Utility: metrics for per-pixel predictions
def compute_pixel_metrics(y_true, y_prob, threshold=0.5, verbose=True):
    # y_true, y_prob shapes: (N, H, W, 1)
    y_true_flat = y_true.reshape(-1)
    y_prob_flat = y_prob.reshape(-1)
    y_pred_flat = (y_prob_flat > threshold).astype(np.uint8)
    precision = precision_score(y_true_flat, y_pred_flat, zero_division=0)
    recall = recall_score(y_true_flat, y_pred_flat, zero_division=0)
    f1 = f1_score(y_true_flat, y_pred_flat, zero_division=0)
    auc = roc_auc_score(y_true_flat, y_prob_flat)
    if verbose:
        print(f"Precision: {precision:.4f}  Recall: {recall:.4f}  F1: {f1:.4f}  AUC: {auc:.4f}")
    return {"precision": precision, "recall": recall, "f1": f1, "auc": auc}

# 7. DNN BASELINE (per-pixel output)
def build_dnn_baseline(ris_size=RIS_SIZE, channels=CHANNELS):
    inp = Input(shape=(ris_size, ris_size, channels))
    x = layers.Flatten()(inp)
    x = Dense(1024, activation='relu')(x)
    x = Dropout(0.4)(x)
    x = Dense(1024, activation='relu')(x)
    x = Dropout(0.3)(x)
    x = Dense(512, activation='relu')(x)
    out = Dense(ris_size * ris_size, activation='sigmoid')(x)  # per-pixel probabilities
    out_reshaped = layers.Reshape((ris_size, ris_size, 1))(out)
    model = Model(inp, out_reshaped, name="DNN_baseline")
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
                  loss='binary_crossentropy', metrics=[tf.keras.metrics.AUC(name='auc')])
    return model

dnn = build_dnn_baseline()
dnn.summary()

# 8. CNN (small U-Net) for per-pixel segmentation
def build_small_unet(input_shape=(RIS_SIZE, RIS_SIZE, CHANNELS)):
    inp = Input(shape=input_shape)

    # Encoder
    c1 = Conv2D(32, 3, padding='same', activation='relu')(inp)
    c1 = BatchNormalization()(c1)
    c1 = Conv2D(32, 3, padding='same', activation='relu')(c1)
    p1 = MaxPooling2D()(c1)

    c2 = Conv2D(64, 3, padding='same', activation='relu')(p1)
    c2 = BatchNormalization()(c2)
    c2 = Conv2D(64, 3, padding='same', activation='relu')(c2)
    p2 = MaxPooling2D()(c2)

    c3 = Conv2D(128, 3, padding='same', activation='relu')(p2)
    c3 = BatchNormalization()(c3)
    c3 = Conv2D(128, 3, padding='same', activation='relu')(c3)

    # Decoder
    u2 = UpSampling2D()(c3)
    m2 = Concatenate()([u2, c2])
    c4 = Conv2D(64, 3, padding='same', activation='relu')(m2)
    c4 = BatchNormalization()(c4)
    c4 = Conv2D(64, 3, padding='same', activation='relu')(c4)

    u1 = UpSampling2D()(c4)
    m1 = Concatenate()([u1, c1])
    c5 = Conv2D(32, 3, padding='same', activation='relu')(m1)
    c5 = BatchNormalization()(c5)
    c5 = Conv2D(32, 3, padding='same', activation='relu')(c5)

    out = Conv2D(1, 1, padding='same', activation='sigmoid')(c5)
    model = Model(inp, out, name="Small_UNet")
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
                  loss='binary_crossentropy', metrics=[tf.keras.metrics.BinaryAccuracy(), tf.keras.metrics.AUC(name='auc')])
    return model

unet = build_small_unet()
unet.summary()

```

```python
# 9. Train both models (use smaller subset if needed to save time)
# For reproducibility, shuffle is False (Keras shuffles by default, but we set seed)
# Use a small validation split; we will keep the test set aside.
# Save best models to disk
os.makedirs('models', exist_ok=True)
dnn_ckpt = ModelCheckpoint('models/dnn_best.h5', save_best_only=True, monitor='val_loss', mode='min')
unet_ckpt = ModelCheckpoint('models/unet_best.h5', save_best_only=True, monitor='val_loss', mode='min')

print("\nTraining DNN baseline...")
history_dnn = dnn.fit(X_train_full, Y_train_full,
                      validation_split=VAL_SPLIT,
                      epochs=EPOCHS,
                      batch_size=BATCH_SIZE,
                      callbacks=[EarlyStopping(patience=8, restore_best_weights=True), ReduceLROnPlateau(patience=4), dnn_ckpt],
                      verbose=2)

```

```python
print("\nTraining U-Net CNN...")
history_unet = unet.fit(X_train_full, Y_train_full,
                        validation_split=VAL_SPLIT,
                        epochs=EPOCHS,
                        batch_size=BATCH_SIZE,
                        callbacks=[EarlyStopping(patience=8, restore_best_weights=True), ReduceLROnPlateau(patience=4), unet_ckpt],
                        verbose=2)

```

```python
# 10. Evaluation on held-out test set
print("\nEvaluating DNN on test set...")
dnn_preds = dnn.predict(X_test, batch_size=128)
compute_pixel_metrics(Y_test, dnn_preds, threshold=0.5)

print("\nEvaluating U-Net on test set...")
unet_preds = unet.predict(X_test, batch_size=128)
compute_pixel_metrics(Y_test, unet_preds, threshold=0.5)

```

```python
# 11. Per-sample visual check: show a few examples (input real channel, imag channel, ground-truth, preds)
def show_examples(X, Y_true, Y_pred, idxs=None, n=3):
    if idxs is None:
        idxs = np.random.choice(X.shape[0], n, replace=False)
    for i in idxs:
        fig, axes = plt.subplots(1, 4, figsize=(12,3))
        axes[0].imshow(X[i,:,:,0], cmap='bwr'); axes[0].set_title('Real channel')
        axes[1].imshow(X[i,:,:,1], cmap='bwr'); axes[1].set_title('Imag channel')
        axes[2].imshow(Y_true[i,:,:,0], cmap='gray'); axes[2].set_title('Ground truth')
        axes[3].imshow(Y_pred[i,:,:,0] > 0.5, cmap='gray'); axes[3].set_title('Pred (th=0.5)')
        plt.show()

print("\nExample results for U-Net (first 3 test samples):")
show_examples(X_test, Y_test, unet_preds, n=3)

```

```python
# 12. Training curves
def plot_history(hist, title_prefix=''):
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(hist.history['loss'], label='train loss')
    plt.plot(hist.history['val_loss'], label='val loss')
    plt.title(title_prefix + ' Loss')
    plt.legend()
    plt.subplot(1,2,2)
    if 'auc' in hist.history:
        plt.plot(hist.history.get('auc'), label='train auc')
        plt.plot(hist.history.get('val_auc'), label='val auc')
        plt.title(title_prefix + ' AUC')
        plt.legend()
    plt.show()

plot_history(history_dnn, 'DNN')
plot_history(history_unet, 'U-Net')

```

```python
# 13. Optional: Simplified end-to-end SINR & sum-rate comparison
# This is a simplified model to measure effect of using predicted mask vs true mask.
# It follows the structure: desired_power = |h_UR^T diag(phi * mask) h_RB|^2 (very simplified)
# For demonstration we simulate random channel vectors and compute SINR ratio.

def simulate_channels(num_experiments=1000, ris_size=RIS_SIZE, num_users=1, seed=SEED):
    np.random.seed(seed)
    # Random UE-to-RIS and RIS-to-BS channels (complex). We'll flatten RIS to vector.
    results = []
    for _ in range(num_experiments):
        # Complex channels: vectors of length M (flattened)
        h_ur = (np.random.normal(0,1,ris_size*ris_size) + 1j*np.random.normal(0,1,ris_size*ris_size))
        h_rb = (np.random.normal(0,1,ris_size*ris_size) + 1j*np.random.normal(0,1,ris_size*ris_size))
        results.append((h_ur, h_rb))
    return results

def compute_gain(h_ur, h_rb, phase_shifts=None, mask=None):
    # h_ur, h_rb: flattened complex vectors length M
    # phase_shifts: vector length M of complex unit-modulus values (we'll assume perfect phases if None)
    # mask: binary vector length M where 1 -> healthy (use phase_shifts), 0 -> failed (zeroed)
    M = h_ur.size
    if mask is None:
        mask = np.ones(M)
    if phase_shifts is None:
        phase_shifts = np.ones(M, dtype=complex)
    # effective element weights: phase_shifts * mask
    w = phase_shifts * mask
    # simple cascaded channel ~ h_rb^T diag(w) h_ur (scalar)
    val = np.vdot(h_rb, w * h_ur)  # conjugate dot product
    gain = np.abs(val)**2
    return gain

# Prepare channel sims and evaluate average gain under different masks
sims = simulate_channels(num_experiments=200)
# flatten the masks for test set samples to compare predicted vs true
true_masks_flat = 1 - Y_test.reshape(Y_test.shape[0], -1)  # 1 means healthy -> we invert because Y==1 is failed
pred_masks_dnn_flat = 1 - dnn_preds.reshape(dnn_preds.shape[0], -1)
pred_masks_unet_flat = 1 - unet_preds.reshape(unet_preds.shape[0], -1)

def average_gain_over_test(sims, mask_flat_arr):
    gains_avg = []
    # Use first len(mask_flat_arr) sims entries
    n = min(len(mask_flat_arr), len(sims))
    for i in range(n):
        h_ur, h_rb = sims[i]
        mask = mask_flat_arr[i]
        # threshold mask to 0/1
        mask_bin = (mask > 0.5).astype(float)
        g = compute_gain(h_ur, h_rb, phase_shifts=None, mask=mask_bin)
        gains_avg.append(g)
    return np.mean(gains_avg)

print("\nComputing simplified average gains (higher is better):")
g_true = average_gain_over_test(sims, true_masks_flat)
g_dnn = average_gain_over_test(sims, pred_masks_dnn_flat)
g_unet = average_gain_over_test(sims, pred_masks_unet_flat)
print(f"Average gain (oracle true mask): {g_true:.4e}")
print(f"Average gain (DNN-predicted mask): {g_dnn:.4e}")
print(f"Average gain (U-Net-predicted mask): {g_unet:.4e}")
print("Relative gain (U-Net / DNN): {:.3f}".format(g_unet / (g_dnn + 1e-12)))

# Optional: compute simple "sum-rate" related metric using Shannon: rate = log2(1 + S/(N+I)) with noise=1
def avg_rate_from_gain(gains, noise_var=1.0):
    return np.mean(np.log2(1 + np.array(gains)/noise_var))

# compute per-sample gains arrays for distributions
def gains_array(sims, mask_flat_arr):
    arr = []
    n = min(len(mask_flat_arr), len(sims))
    for i in range(n):
        h_ur, h_rb = sims[i]
        mask = mask_flat_arr[i]
        g = compute_gain(h_ur, h_rb, mask=(mask>0.5).astype(float))
        arr.append(g)
    return np.array(arr)

gains_true_arr = gains_array(sims, true_masks_flat)
gains_dnn_arr = gains_array(sims, pred_masks_dnn_flat)
gains_unet_arr = gains_array(sims, pred_masks_unet_flat)

print("\nAvg rate (oracle):", avg_rate_from_gain(gains_true_arr))
print("Avg rate (DNN):", avg_rate_from_gain(gains_dnn_arr))
print("Avg rate (U-Net):", avg_rate_from_gain(gains_unet_arr))

```
